```python
methods = [brute.Method.BRUTE_FORCE, brute.Method.SMART_FORCE]

data = []

while True:
	for num_vertices in [100]:
		for num_edges in range(10, 200,10):
			graph = gen_num_edges(num_vertices, num_edges)

			for method in methods:

				elapsed = brute.solve(graph, method)
				print(f"{method.name}(v:{num_vertices},e:{num_edges}): {elapsed}")
				data.append({
					"method": method.name,
					"elapsed": elapsed,
					"num_edges": num_edges,
					"num_vertices": num_vertices,
				})
	break

df = pd.DataFrame(data)
df
```
